# **Series**

* A Pandas Series is like a ***column in a table***.
* It is a ***one-dimensional array holding data of any type.***

**1. Creating a Series using List**

```python
# Create a simple Pandas Series from a list:
import pandas as pd
a = [1, 7, 2]
myvar = pd.Series(a)
print(myvar)
```

**2. Labels in Pandas Series**
* If nothing else is specified, the values are labeled with their index number. First value has index 0, second value has index 1 etc.
* This label can be used to access a specified value

```python
# Return the first value of the Series:
print(myvar[0])
```

**3. Create Labels for Pandas Series**
* With the index argument, you can name your own labels.

```python
# Create you own labels:
import pandas as pd
a = [1, 7, 2]
myvar = pd.Series(a, index = ["x", "y", "z"])
print(myvar)
```

**Note :** When you have created labels, you can access an item by referring to the label.

```python
# Return the value of "y":
print(myvar["y"])
```

**4. Key/Value Objects as Series**
* Dictionary can be use for creating a Series.

```python
# Create a simple Pandas Series from a dictionary:
import pandas as pd
calories = {"day1": 420, "day2": 380, "day3": 390}
myvar = pd.Series(calories)
print(myvar)
```

**Note:**
* The keys of the dictionary become the labels.
* To select only some of the items in the dictionary, use the index argument and specify only the items you want to include in the Series

```python
#Create a Series using only data from "day1" and "day2":
import pandas as pd
calories = {"day1": 420, "day2": 380, "day3": 390}
myvar = pd.Series(calories, index = ["day1", "day2"])
print(myvar)
```

# **Data Frames**

* ***Data sets*** in Pandas are usually ***multi-dimensional tables, called DataFrames.***
* Series is like a column, a DataFrame is the whole table.

**1. Creating a DataFrame**

```python
#Create a DataFrame from two Series:
import pandas as pd
data = {
  "calories": [420, 380, 390],
  "duration": [50, 40, 45]
}
df = pd.DataFrame(data)
print(df)
```

**2. Locate a row**
* Pandas use the ***loc*** attribute to return one or more specified row(s)

```python
#refer to the row index:
print(df.loc[0])
```

```python
# Return row 0 and 1 using a list of indexes:
print(df.loc[[0, 1]])
```

**Note:** When using [], the result is a Pandas DataFrame.

```python
print(df.loc[[]])
```

**3. Named Indexes**
* **index** argument used to name the indexes.

```python
#Add a list of names to give each row a name:
import pandas as pd
data = {
  "calories": [420, 380, 390],
  "duration": [50, 40, 45]
}
df = pd.DataFrame(data, index = ["day1", "day2", "day3"])
print(df)
```

**4. Locate Named Indexes**
* Use the named index in the **loc** attribute to return the specified row(s). 

```python
#refer to the named index to Return "day2":
print(df.loc["day2"])
```

```python
# 1. Rescaling(Min-Max Normalization)
## -- simplest type of normalization where the features are rescaled to the range [0, 1]

# Min Max Normalization using MinMaxScaler in sklearn library
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()
TotalPayReshaped = df_fs.TotalPay.values.reshape(-1,1)
df_fs.TotalPay = scaler.fit_transform(TotalPayReshaped)

# minimum value of TotalPay is 0 and the maximum is 1.
df_fs.describe()
```

```python
# 2. Standardization (Z-score normalization)
## -- Here feature is scaled such that it’s mean is zero and variance is one.

from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()

TotalPayReshaped = df_fs.TotalPay.values.reshape(-1,1)
df_fs.TotalPay = scaler.fit_transform(TotalPayReshaped)

# Here mean of TotalPay is close to 0 and the standard deviation is 1.
df_fs.describe()
```

```python
# Convert text to lowercase
input_str = 'The 5 biggest countries by population in 2017 are China, India, United States, Indonesia, and Brazil.'
input_str = input_str.lower()
print(input_str)
```

```python
# Remove numbers if they are not relevant to your analyses. 
# Usually, regular expressions are used to remove numbers.
import re
input_str = 'Box A contains 3 red and 5 white balls, while Box B contains 4 red and 2 blue balls.'
result = re.sub(r'\d+', '', input_str)
print(result)
```

```python
# Remove punctuation:  [!”#$%&’()*+,-./:;<=>?@[\]^_`{|}~]:
punctuations = '''!()-[]{};:'"\,<>./?@#$%^&*_~'''
 
inp_str = input("Enter a string: ")
 
no_punc = ""
for char in inp_str:
   if char not in punctuations:
       no_punc = no_punc + char
 
print("Punctuation Free String: ",no_punc)
```

```python
# Remove whitespaces
# To remove leading and ending spaces, you can use the strip() function
input_str = ' \t a string example\t '
input_str = input_str.strip()
input_str
```
